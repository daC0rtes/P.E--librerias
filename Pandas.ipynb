```python
# filtraremos y modificaremos aquellos registros que contengan Mega en el nombre
# df.loc[~df['Name'].str.contains('Mega')]
#si usamos el simbolo ~ veremos que nuestra data se "limpiara" de los nombres que contengan Mega
df.loc[df['Type 1'].str.contains('Fire|Grass', regex =True)]
"""
lo anterior se puede llevar a cabo de la siguiente manera
import re
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I,regex =True)]
"""
#si queremos los registro de Name que contengan  pi se hace de la siguiente manera
import re
df.loc[df['Name'].str.contains('pi[a-z]*', flags = re.I,regex =True)]
#si queremos los registro de Name que empiezen por  pi se hace de la siguiente manera
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I,regex =True)]
```

Estadisticas "Agragadas"

Aggregate Statistics (Groupby)

```python
df = pd.read_csv("/content/modificado.csv")
df.groupby(['Type 1']).mean().sort_values('Defense',ascending=False)
#de donde podemos observar que los pokemones de hierro son los que tienen mas defensa
# y asi usando sortby y groupby podemos acceder a varios datos descriptivos de manera rapida y facil
```

```python
# manera 1
df.groupby(['Type 1']).count()
#manera 2 mucho mas facil de leer y comoda
df['count'] = 1
df.groupby(['Type 1']).count()['count']

#para subconjuntos de interes filtramos multiples veces

df.groupby(['Type 1', 'Type 2']).count()['count']

```
